```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv(r'C:\Users\Tobi\Desktop\BI\bluegills.txt', sep="\t") ##Adding the command sep"\t" because the data is seperated by tabs and not ','
```

```python
list(df)
```

```python
df.isnull().sum()
```

```python
df.describe().round(2)
```

```python
df['age'].value_counts().plot(kind='bar')

plt.title('Bluegills age')
plt.ylabel('Amount')
plt.show()
```

```python
df['age'].value_counts().plot(kind='pie')

plt.title('Bluegills age distribution')
plt.ylabel("")
plt.show()
```

```python
df['length'].value_counts().plot(kind='bar')

plt.title('Bluegills')
plt.ylabel('Length in mm')
plt.show()
```

```python
df['length'].value_counts().plot(kind='pie')

plt.title('Bluegills distribution of length in mm')
plt.ylabel("")
plt.show()
```
