```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

dfpen = pd.read_csv(r'C:\Users\Tobi\Desktop\BI\penguins.csv', index_col=None, na_values=['NA'])
dfpen.shape
```

```python
dfpen.describe().round(2)
```

```python
dfpen.isnull().sum()
```

Using the mean data for the missing data points in the set for culmen, flippers and mass

```python
mean_val = dfpen["culmen_length_mm"].mean()
dfpen["culmen_length_mm"] = dfpen["culmen_length_mm"].fillna(mean_val)
```

```python
mean_val = dfpen["culmen_depth_mm"].mean()
dfpen["culmen_depth_mm"] = dfpen["culmen_depth_mm"].fillna(mean_val)
```

```python
mean_val = dfpen["flipper_length_mm"].mean()
dfpen["flipper_length_mm"] = dfpen["flipper_length_mm"].fillna(mean_val)
```

```python
mean_val = dfpen["body_mass_g"].mean()
dfpen["body_mass_g"] = dfpen["body_mass_g"].fillna(mean_val)
```

Setting the missing data for genders as 'NA'

```python
dfpen["sex"] = dfpen["sex"].fillna("NA")
```

```python
dfpen['species'].value_counts().plot(kind='bar')

plt.title('Species of penguins')
plt.ylabel('Penguins')
plt.show()
```

```python
dfpen['sex'].value_counts().plot(kind='bar')

plt.title('Gender')
plt.ylabel('Penguins')
plt.show()
```

```python
dfpen['island'].value_counts().plot(kind='bar')

plt.title('Islands')
plt.ylabel('Penguins')
plt.show()
```

```python
stats = dfpen['culmen_length_mm'].describe()
summary = stats[['min', 'mean', 'max']]
summary.plot(kind='bar')
plt.title('Length of culmen')
plt.ylabel('Length (mm)')
plt.show()
```

```python
stats = dfpen['culmen_depth_mm'].describe()
summary = stats[['min', 'mean', 'max']]
summary.plot(kind='bar')
plt.title('Depth of culmen')
plt.ylabel('Length (mm)')
plt.show()
```

```python
stats = dfpen['flipper_length_mm'].describe()
summary = stats[['min', 'mean', 'max']]
summary.plot(kind='bar')
plt.title('Length of flipper')
plt.ylabel('Length (mm)')
plt.show()
```

```python
stats = dfpen['body_mass_g'].describe()
summary = stats[['min', 'mean', 'max']]
summary.plot(kind='bar')
plt.title('Weight of penguins')
plt.ylabel('Weight (gram)')
plt.show()
```
