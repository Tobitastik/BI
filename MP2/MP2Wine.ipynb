```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

white = pd.read_excel("winequality-white.xlsx", header=1)
red = pd.read_excel("winequality-red.xlsx", header=1)
red['type'] = 'red'
white['type'] = 'white'

df_wine = pd.concat([white, red], ignore_index=True)

df_wine.isnull().sum()
```

Before look at the dataset I would like to talk about the data set. The collection of white wine data are more than three times larger than the red wine and while the dataset of 1699 red wines are large this might scue the data. Also the data was collected from 2004 to 2007 within a single region in Portgual, this could affect the data if the wine harvest have either been really bad or good.

```python
sns.countplot(data=df_wine, x="type", hue="type")
plt.title("Count of red and white wine in the dataset")
plt.show()
```

```python
df_red = df_wine[df_wine["type"] == "red"]
df_white = df_wine[df_wine["type"] == "white"]

corr_red = df_red.drop(columns="type").corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_red, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - Red Wine")
plt.show()

corr_white = df_white.drop(columns="type").corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_white, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - White Wine")
plt.show()
```

Bins

```python
df_wine["pH_bin_5"] = pd.cut(df_wine["pH"], bins=5)

counts_5 = df_wine.groupby("pH_bin_5", observed=True).size()


bin_widths_5 = df_wine["pH_bin_5"].cat.categories.map(lambda x: x.right - x.left)

density_5 = counts_5 / bin_widths_5

print("Density with 5 bins:")
print(density_5)

print("\nHighest density (5 bins):", density_5.idxmax(), "with density", density_5.max())
```

```python
df_wine["pH_bin_10"] = pd.cut(df_wine["pH"], bins=10)

counts_10 = df_wine.groupby("pH_bin_10", observed=True).size()

bin_widths_10 = df_wine["pH_bin_10"].cat.categories.map(lambda x: x.right - x.left)

density_10 = counts_10 / bin_widths_10

print("Density with 10 bins:")
print(density_10)

print("\nHighest density (10 bins):", density_10.idxmax(), "with density", density_10.max())
```

```python
numeric_df = df_wine.drop(columns=["type", "pH_bin_5", "pH_bin_10"], errors="ignore")

corr = numeric_df.corr()

quality_corr = corr["quality"].sort_values(ascending=False)
print("Correlation of wine attributes with quality:\n")
print(quality_corr)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Wine Attributes")
plt.show()
```

From the data above, alcohol appear to be the strongest factor for the quality of the wine while pH, residual sugar, sulphates and sulfur dioxide seems to be the weakest impact for the quality with correaltion close to zero.
However alcohol alone likely doesn't play too large a role given its correaltion of only 0.44.
We have yet to find the single messsureable attribute which make a wine truly great.
It may also be that wine quality is the result of a combination of many factors rather than any single one.

```python
corr_red = df_red.drop(columns="type").corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_red, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - Red Wine")
plt.show()

corr_white = df_white.drop(columns="type").corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_white, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - White Wine")
plt.show()
```

By splitting the two types of wine, we can see slight differences in how quality correlates with certain factors for example, sulphates show a stronger influence in red wine compared to white wine, while density is more influential in white wine than in red. However, these correlations are still far too weak to be considered truly significant.
